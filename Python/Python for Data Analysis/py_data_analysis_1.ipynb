```python
import re
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]', '', value)
        value = value.title()
        result.append(value)
    return result
```

```python
states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda', 'south carolina##', 'West virginia?']
result = clean_strings(states)
print(result)
```

```python
def remove_punctuation(value):
    return re.sub('[!#?]', '', value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_strings(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
```

```python
result2 = clean_strings(states, clean_ops)
print(result2)
```

```python
for x in map(remove_punctuation, states):
    print(x)
```

```python
all_genres = []

for gen in movies.genres:
    all_genres.extend(gen.split('|'))
    
genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))

dummies = pd.DataFrame(zero_matrix, columns=genres)

for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
    
dummies
```

## 10.3 Apply: General split-apply-combine

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]
```

```python
top(tips)
```

```python
tips.groupby('smoker').apply(top)
```

What has happened here? The top function is called on each row group from the
DataFrame, and then the results are glued together using pandas.concat, labeling the
pieces with the group names. The result therefore has a hierarchical index whose
inner level contains index values from the original DataFrame.

```python
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
tips.groupby('smoker', group_keys=False).apply(top)
```
